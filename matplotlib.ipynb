```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotnine as p9
```

```python
x = np.linspace(0, 5, 100 )  # Sample data, 100 points between 0 and 5.

fig, ax = plt.subplots(figsize=(10, 6))

ax.grid(True) # Add a grid

ax.plot(x, np.sin(x), label='Sine function $\sin(x)$')  # Plot some data on the axes. Can use LaTeX notation.
ax.plot(x, x, label='linear $x$')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic $x^{3}$')  # ... and some more.
ax.plot(x, x-x**3/6.0, linestyle='dashed', label='$x - x^{3} / 6$')

ax.set_xlim([0, np.pi/2]) # We select limits for plotting
ax.set_ylim([0, 1.5])

ax.set_xticks([0, 0.125*np.pi, 0.25*np.pi, 0.375*np.pi, np.pi/2]) # values
ax.set_xticklabels(['$0$', '$\pi/8$','$\pi/4$','$3 \pi/8$','$\pi/2$']) # labels

ax.set_xlabel('$x$')  # Add an x-label to the axes.
ax.set_ylabel('$y = f(x)$')  # Add a y-label to the axes.
ax.set_title("Simple function plots")  # Add a title to the axes.

ax.legend();  # Add a legend.
```

```python
x1 = np.linspace(0.0, 5.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
x2 = np.linspace(0.0, 2.0)
y2 = np.cos(2 * np.pi * x2)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 6))

fig.suptitle('Making multiple plots') # General title

ax1.grid(True) # Add a grid to first plot
ax1.plot(x1, y1, 'o-', label='$\cos(2 \pi x) \exp(-x)$')
ax1.set_ylabel('Damped oscillation')

ax2.grid(True) # Add a grid to second plot
ax2.plot(x2, y2, '.-')
ax2.set_xlabel('time (s)')
ax2.set_ylabel('Undamped')

plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.grid(True)
plt.plot(x1, y1, 'o-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)
plt.grid(True)
plt.plot(x2, y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

from mpl_toolkits.mplot3d.axes3d import get_test_data
```

```python
fig = plt.figure(figsize=plt.figaspect(0.333))

ax = fig.add_subplot(1, 2, 1, projection='3d')

X = np.arange(-5, 5, 0.125)
Y = np.arange(-5, 5, 0.125)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
ax.set_zlim(-1.01, 1.01)
fig.colorbar(surf, shrink=0.5, aspect=10)
```
